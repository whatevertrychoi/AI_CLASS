```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

# 1ï¸âƒ£ ë°ì´í„° ë¡œë“œ
df = pd.read_csv("C:/Users/Minseok/AI_CLASS/wine.csv")

# 2ï¸âƒ£ ì…ë ¥(X)ê³¼ ì¶œë ¥(y) ë¶„ë¦¬
X = df.drop(columns=['Wine'])  # ì…ë ¥ ë°ì´í„° (13ê°œ íŠ¹ì§•)
y = df['Wine'] - 1  # í´ë˜ìŠ¤ ê°’ (1ë¶€í„° ì‹œì‘í•˜ëŠ” ê²ƒì„ 0ë¶€í„° ì‹œì‘í•˜ë„ë¡ ë³€í™˜)

# 3ï¸âƒ£ ë°ì´í„° í‘œì¤€í™” (ëª¨ë¸ í•™ìŠµì„ ìœ„í•´ ì •ê·œí™”)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 4ï¸âƒ£ ë°ì´í„° ë¶„í•  (í•™ìŠµ 80% / í…ŒìŠ¤íŠ¸ 20%)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 5ï¸âƒ£ ì›-í•« ì¸ì½”ë”© ìˆ˜í–‰
encoder = OneHotEncoder(sparse_output=False)
y_train_onehot = encoder.fit_transform(y_train.values.reshape(-1, 1))
y_test_onehot = encoder.transform(y_test.values.reshape(-1, 1))

# --------------------- [ì›-í•« ì¸ì½”ë”© ì „í›„ ë¹„êµ] ---------------------
models = {
    "Logistic Regression": LogisticRegression(max_iter=200),
    "SVM": SVC(kernel='linear', probability=True),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "MLP (Neural Network)": MLPClassifier(hidden_layer_sizes=(10, 8, 6), activation='relu', max_iter=500, random_state=42)
}

accuracy_results = {}

# 6ï¸âƒ£ ê° ëª¨ë¸ë³„ í•™ìŠµ ë° í‰ê°€ (ì›-í•« ì¸ì½”ë”© ì „)
print("ğŸ”¹ ë¨¸ì‹ ëŸ¬ë‹ ëª¨ë¸ í•™ìŠµ ì¤‘... (ì›-í•« ì¸ì½”ë”© ì „)")
for model_name, model in models.items():
    model.fit(X_train, y_train)  # ëª¨ë¸ í•™ìŠµ
    y_pred = model.predict(X_test)  # ì˜ˆì¸¡
    acc = accuracy_score(y_test, y_pred)  # ì •í™•ë„ ê³„ì‚°
    accuracy_results[model_name] = {"No One-Hot": acc}
    print(f"âœ… {model_name} ì •í™•ë„ (ì›-í•« ì¸ì½”ë”© ì „): {acc:.4f}")

# 7ï¸âƒ£ ê° ëª¨ë¸ë³„ í•™ìŠµ ë° í‰ê°€ (ì›-í•« ì¸ì½”ë”© í›„)
print("\nğŸ”¹ ë¨¸ì‹ ëŸ¬ë‹ ëª¨ë¸ í•™ìŠµ ì¤‘... (ì›-í•« ì¸ì½”ë”© í›„)")
for model_name, model in models.items():
    model.fit(X_train, np.argmax(y_train_onehot, axis=1))  # ëª¨ë¸ í•™ìŠµ (ë¼ë²¨ì„ ë‹¤ì‹œ ì •ìˆ˜í˜•ìœ¼ë¡œ ë³€í™˜)
    y_pred_onehot = model.predict(X_test)  # ì˜ˆì¸¡
    acc_onehot = accuracy_score(np.argmax(y_test_onehot, axis=1), y_pred_onehot)  # ì •í™•ë„ ê³„ì‚°
    accuracy_results[model_name]["One-Hot"] = acc_onehot
    print(f"âœ… {model_name} ì •í™•ë„ (ì›-í•« ì¸ì½”ë”© í›„): {acc_onehot:.4f}")

# 8ï¸âƒ£ ìµœì¢… ê²°ê³¼ ì¶œë ¥
print("\nğŸ“Š ëª¨ë¸ë³„ ì •í™•ë„ ë¹„êµ ğŸ“Š")
for model_name, acc in accuracy_results.items():
    print(f"ğŸ”¹ {model_name}: ì›-í•« ì¸ì½”ë”© ì „={acc['No One-Hot']:.4f}, ì›-í•« ì¸ì½”ë”© í›„={acc['One-Hot']:.4f}")
```
