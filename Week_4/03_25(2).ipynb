```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

# 1️⃣ 데이터 로드
df = pd.read_csv("C:/Users/Minseok/AI_CLASS/wine.csv")

# 2️⃣ 입력(X)과 출력(y) 분리
X = df.drop(columns=['Wine'])  # 입력 데이터 (13개 특징)
y = df['Wine'] - 1  # 클래스 값 (1부터 시작하는 것을 0부터 시작하도록 변환)

# 3️⃣ 데이터 표준화 (모델 학습을 위해 정규화)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 4️⃣ 데이터 분할 (학습 80% / 테스트 20%)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 5️⃣ 원-핫 인코딩 수행
encoder = OneHotEncoder(sparse_output=False)
y_train_onehot = encoder.fit_transform(y_train.values.reshape(-1, 1))
y_test_onehot = encoder.transform(y_test.values.reshape(-1, 1))

# --------------------- [원-핫 인코딩 전후 비교] ---------------------
models = {
    "Logistic Regression": LogisticRegression(max_iter=200),
    "SVM": SVC(kernel='linear', probability=True),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "MLP (Neural Network)": MLPClassifier(hidden_layer_sizes=(10, 8, 6), activation='relu', max_iter=500, random_state=42)
}

accuracy_results = {}

# 6️⃣ 각 모델별 학습 및 평가 (원-핫 인코딩 전)
print("🔹 머신러닝 모델 학습 중... (원-핫 인코딩 전)")
for model_name, model in models.items():
    model.fit(X_train, y_train)  # 모델 학습
    y_pred = model.predict(X_test)  # 예측
    acc = accuracy_score(y_test, y_pred)  # 정확도 계산
    accuracy_results[model_name] = {"No One-Hot": acc}
    print(f"✅ {model_name} 정확도 (원-핫 인코딩 전): {acc:.4f}")

# 7️⃣ 각 모델별 학습 및 평가 (원-핫 인코딩 후)
print("\n🔹 머신러닝 모델 학습 중... (원-핫 인코딩 후)")
for model_name, model in models.items():
    model.fit(X_train, np.argmax(y_train_onehot, axis=1))  # 모델 학습 (라벨을 다시 정수형으로 변환)
    y_pred_onehot = model.predict(X_test)  # 예측
    acc_onehot = accuracy_score(np.argmax(y_test_onehot, axis=1), y_pred_onehot)  # 정확도 계산
    accuracy_results[model_name]["One-Hot"] = acc_onehot
    print(f"✅ {model_name} 정확도 (원-핫 인코딩 후): {acc_onehot:.4f}")

# 8️⃣ 최종 결과 출력
print("\n📊 모델별 정확도 비교 📊")
for model_name, acc in accuracy_results.items():
    print(f"🔹 {model_name}: 원-핫 인코딩 전={acc['No One-Hot']:.4f}, 원-핫 인코딩 후={acc['One-Hot']:.4f}")
```
