```python
import matplotlib.pyplot as plt
from sklearn import datasets, metrics
from sklearn.model_selection import train_test_split

digits = datasets.load_digits()
plt.imshow(digits.images[0], cmap = plt.cm.gray_r, interpolation = 'nearest')
```

```python
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, digits.target, test_size = 0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 6) 
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)

scores = metrics.accuracy_score(y_test, y_pred)
scores
```

```python

from sklearn.linear_model import LogisticRegression

# 모델 생성 및 학습
model_lr = LogisticRegression(max_iter=1000)
model_lr.fit(X_train, y_train)

# 예측 및 평가
y_pred_lr = model_lr.predict(X_test)
acc_lr = metrics.accuracy_score(y_test, y_pred_lr)

print("\n🏆 [로지스틱 회귀]")
print(f"✅ 정확도: {acc_lr:.4f}")
```

```python

from sklearn.tree import DecisionTreeClassifier

# 모델 생성 및 학습
model_dt = DecisionTreeClassifier(random_state=42)
model_dt.fit(X_train, y_train)

# 예측 및 평가
y_pred_dt = model_dt.predict(X_test)
acc_dt = metrics.accuracy_score(y_test, y_pred_dt)

print("\n🌳 [의사결정나무]")
print(f"✅ 정확도: {acc_dt:.4f}")
```

```python

from sklearn.ensemble import RandomForestClassifier

# 모델 생성 및 학습
model_rf = RandomForestClassifier(n_estimators=100, random_state=42)
model_rf.fit(X_train, y_train)

# 예측 및 평가
y_pred_rf = model_rf.predict(X_test)
acc_rf = metrics.accuracy_score(y_test, y_pred_rf)

print("\n🌲 [랜덤 포레스트]")
print(f"✅ 정확도: {acc_rf:.4f}")
```

```python

from sklearn.svm import SVC

# 모델 생성 및 학습
model_svm = SVC(kernel="linear")
model_svm.fit(X_train, y_train)

# 예측 및 평가
y_pred_svm = model_svm.predict(X_test)
acc_svm = metrics.accuracy_score(y_test, y_pred_svm)

print("\n🎯 [서포트 벡터 머신]")
print(f"✅ 정확도: {acc_svm:.4f}")
```
