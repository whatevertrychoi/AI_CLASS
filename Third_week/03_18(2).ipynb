```python
# 1. 데이터 로드
import pandas as pd
file_path = "C:/Users/Minseok/AI_CLASS/abalone.csv" 
df = pd.read_csv(file_path) #csv_file load
df.head()
# 2. 결측치 확인
print("🔍 [결측치 개수 확인]")
print(df.isnull().sum())  # 각 컬럼별 결측치 개수 출력

# 3. 레이블 확인 (Rings)
print("\n🎯 [타겟 레이블 확인]")
print(df['Rings'].value_counts())  

# 4. 문자 데이터를 숫자로 변환 (변환 전/후 비교)
from sklearn.preprocessing import LabelEncoder
print("\n🔄 [변환 전 데이터 확인]")
print(df.head())  # 변환 전 데이터 확인

label_encoders = {}  # 각 컬럼별 LabelEncoder 저장용
for col in df.select_dtypes(include=['object']).columns:  # 문자열 데이터만 변환
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le  # 인코더 저장

print("\n🔄 [변환 후 데이터 확인]")
print(df.head())  # 변환 후 데이터 확인

#5.데이터 분할 (훈련 & 테스트)
import numpy as np
X = df.drop(columns=['Rings'])  # 입력 변수
y = df['Rings']  # 타겟 변수

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(X,
                                                  y,
                                                  test_size=0.2,
                                                  shuffle=True,
                                                  random_state=12)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(X_train, y_train)
ly_preds = model.predict(X_test)

print('평균제곱근오차', mean_squared_error(ly_preds, y_test))

def mse_np(actual, predicted):
    return np.mean((np.array(actual) - np.array(predicted)) ** 2)

print('평균제곱근오차', mse_np(ly_preds, y_test))

def mse(actual, predicted):
    sum_square_error = sum((a - p) ** 2 for a, p in zip(actual, predicted))
    mean_square_error = sum_square_error / len(actual)
    return mean_square_error

print('평균제곱근오차', mse(ly_preds, y_test))

from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error

dy_preds = model.predict(X_test)
print("DecisionTreeRegressor")
print('평균제곱근오차', mean_squared_error(dy_preds, y_test))

from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error

ry_preds = model.predict(X_test)
print("RandomForestRegressor")
print('평균제곱근오차', mean_squared_error(ry_preds, y_test))
from sklearn.svm import SVR

model = SVR(kernel='linear')
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error

ry_preds = model.predict(X_test)
print("SVR")
print('평균제곱근오차', mean_squared_error(ry_preds, y_test))
```
