```python
# 1. ë°ì´í„° ë¡œë“œ
import pandas as pd
file_path = "C:/Users/Minseok/AI_CLASS/abalone.csv" 
df = pd.read_csv(file_path) #csv_file load
df.head()
# 2. ê²°ì¸¡ì¹˜ í™•ì¸
print("ğŸ” [ê²°ì¸¡ì¹˜ ê°œìˆ˜ í™•ì¸]")
print(df.isnull().sum())  # ê° ì»¬ëŸ¼ë³„ ê²°ì¸¡ì¹˜ ê°œìˆ˜ ì¶œë ¥

# 3. ë ˆì´ë¸” í™•ì¸ (Rings)
print("\nğŸ¯ [íƒ€ê²Ÿ ë ˆì´ë¸” í™•ì¸]")
print(df['Rings'].value_counts())  

# 4. ë¬¸ì ë°ì´í„°ë¥¼ ìˆ«ìë¡œ ë³€í™˜ (ë³€í™˜ ì „/í›„ ë¹„êµ)
from sklearn.preprocessing import LabelEncoder
print("\nğŸ”„ [ë³€í™˜ ì „ ë°ì´í„° í™•ì¸]")
print(df.head())  # ë³€í™˜ ì „ ë°ì´í„° í™•ì¸

label_encoders = {}  # ê° ì»¬ëŸ¼ë³„ LabelEncoder ì €ì¥ìš©
for col in df.select_dtypes(include=['object']).columns:  # ë¬¸ìì—´ ë°ì´í„°ë§Œ ë³€í™˜
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le  # ì¸ì½”ë” ì €ì¥

print("\nğŸ”„ [ë³€í™˜ í›„ ë°ì´í„° í™•ì¸]")
print(df.head())  # ë³€í™˜ í›„ ë°ì´í„° í™•ì¸

#5.ë°ì´í„° ë¶„í•  (í›ˆë ¨ & í…ŒìŠ¤íŠ¸)
import numpy as np
X = df.drop(columns=['Rings'])  # ì…ë ¥ ë³€ìˆ˜
y = df['Rings']  # íƒ€ê²Ÿ ë³€ìˆ˜

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(X,
                                                  y,
                                                  test_size=0.2,
                                                  shuffle=True,
                                                  random_state=12)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(X_train, y_train)
ly_preds = model.predict(X_test)

print('í‰ê· ì œê³±ê·¼ì˜¤ì°¨', mean_squared_error(ly_preds, y_test))

def mse_np(actual, predicted):
    return np.mean((np.array(actual) - np.array(predicted)) ** 2)

print('í‰ê· ì œê³±ê·¼ì˜¤ì°¨', mse_np(ly_preds, y_test))

def mse(actual, predicted):
    sum_square_error = sum((a - p) ** 2 for a, p in zip(actual, predicted))
    mean_square_error = sum_square_error / len(actual)
    return mean_square_error

print('í‰ê· ì œê³±ê·¼ì˜¤ì°¨', mse(ly_preds, y_test))

from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error

dy_preds = model.predict(X_test)
print("DecisionTreeRegressor")
print('í‰ê· ì œê³±ê·¼ì˜¤ì°¨', mean_squared_error(dy_preds, y_test))

from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error

ry_preds = model.predict(X_test)
print("RandomForestRegressor")
print('í‰ê· ì œê³±ê·¼ì˜¤ì°¨', mean_squared_error(ry_preds, y_test))
from sklearn.svm import SVR

model = SVR(kernel='linear')
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error

ry_preds = model.predict(X_test)
print("SVR")
print('í‰ê· ì œê³±ê·¼ì˜¤ì°¨', mean_squared_error(ry_preds, y_test))
```
